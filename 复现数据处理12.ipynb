### 合并文件

```python
#将所有文件合并的同一个csv中
import os
import pandas as pd
import glob
csv_list = glob.glob('*.csv') #查看同文件夹下的csv文件数
print(u'共发现%s个CSV文件'% len(csv_list))
print(u'正在处理............')
dfa=pd.DataFrame()
#加快读取速度,分两次读，11年到12年合起来，13年合起来，合并之前删除无用项
for i in csv_list:
    f=open(i)
    data=pd.read_csv(f)
    data.drop(['Money_Fee','Money_Fee_Rate','Money_Fee_JPY','Bitcoin_Fee','Bitcoin_Fee_JPY','Money',],axis=1,inplace=True)
    dfa=pd.concat([dfa,data])
dfa.drop('Japan','User_Id_Hash','User','User_State','User_Country'],axis=1,inplace=True)#去掉一些有些表没有的列，而有些表有
```

## 处理数据

```python
import pandas as pd
# f = open('12.csv')
df12 = dfb12
#日期、用户ID、类型和比特币删除重复行
df12=df12.drop_duplicates(['Date','Type','User_Id','Bitcoins']) 
#按Trade_Id分组，选择既有buy也有sell的交易
grouped12=df12.pivot_table('Date',index='Trade_Id',columns='Type',aggfunc=len) 
grouped12=grouped12[(grouped12['buy']==1) & (grouped12['sell']==1)]
df结束=df12[df12["Trade_Id"].isin(grouped12.index)]
```

```python
df_new =  df结束.reset_index(drop=True)
df_new
```
