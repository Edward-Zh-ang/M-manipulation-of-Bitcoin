```python
# 获取Mtgox USD/BTC 价格
mtgoxUSD=quandl.get('BCHARTS/MTGOXUSD',start_date='2011-04-01',end_date='2011-12-31')
mtgoxUSD
```

```python
data=pd.read_csv('2011处理完毕.csv')
data['Price']=data['Money']/data['Bitcoins']
data['No']=data['Price']
data['No']=data['No'].index.values.tolist()
data['Date']=pd.to_datetime(data['Date'].astype('datetime64[D]',format ='%Y-%m-%d'))
data=data.set_index(['Date','No'])
data.head()
```

```python
tranverseDay = datetime(2011,4,1)
j=0
k=0
EHA=pd.DataFrame(index=(['Date']), columns=['Trade_Id','User_Id','Type','Bitcoins','Money','Money_JPY','Price'])
ELA=pd.DataFrame(index=(['Date']), columns=['Trade_Id','User_Id','Type','Bitcoins','Money','Money_JPY','Price'])
NMA=pd.DataFrame(index=(['Date']), columns=['Trade_Id','User_Id','Type','Bitcoins','Money','Money_JPY','Price'])
while tranverseDay.year==2011:
    for i in range(0,len(data.loc[tranverseDay])):
        a=data.loc[tranverseDay,k]['Price']
        b=mtgoxUSD.loc[tranverseDay]['High']
        c=mtgoxUSD.loc[tranverseDay]['Low']
        if a>b:
            EHA.loc[tranverseDay]=data.loc[tranverseDay,k]
            j+=1
        elif a<c:
            ELA.loc[tranverseDay]=data.loc[tranverseDay,k]
        else:
            NMA.loc[tranverseDay]=data.loc[tranverseDay,k]
        k+=1
    tranverseDay = tranverseDay + timedelta(days=1)
    print(tranverseDay)
```
